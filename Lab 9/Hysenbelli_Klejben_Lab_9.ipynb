```python
def dailyopenclose(ticker):
    genius = requests.get(f"https://api.polygon.io/v1/open-close/{ticker}/2023-01-09?adjusted=true&apiKey={key}")
    json_file = genius.json()
    return json_file
    
dailyopenclose("AMZN")
```

```python
def previousclose(ticker): 
    genius = requests.get(f"https://api.polygon.io/v2/aggs/ticker/{ticker}/prev?adjusted=true&apiKey={key}")
    json_file = genius.json()
    return json_file

previousclose("AMZN")
```

```python
def news(ticker): 
    genius = requests.get(f"https://api.polygon.io/v2/reference/news?ticker={ticker}&apiKey={key}")
    json_file = genius.json()
    return json_file

news("AMZN")
```

```python
def status(): 
    genius = requests.get(f"https://api.polygon.io/v1/marketstatus/now?apiKey={key}")
    json_file = genius.json()
    return json_file

status()
```

```python
def details(ticker): 
    genius = requests.get(f"https://api.polygon.io/v3/reference/tickers/{ticker}?apiKey={key}")
    json_file = genius.json()
    return json_file

details("AMZN")
```
