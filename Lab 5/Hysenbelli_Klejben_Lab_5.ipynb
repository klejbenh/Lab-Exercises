# Exercise 2

```python
driver = webdriver.Chrome(service=PATH) 
driver.get("https://hoopshype.com/")

link = driver.find_element("link text", "SALARIES") 
link.click()
link = driver.find_element("link text", "PLAYER SALARIES")
link.click()

csv_file = open("Salaries", "w", newline="", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Name", "2022/2023", "2023/2024", "2024/2025", "2025/2026"]) 

try:
    main = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.TAG_NAME, "table"))
    )
    
    rows = main.find_elements(By.TAG_NAME, "tr")
    for row in rows[1:]:
        row_text = row.text
        cols = row_text.split('$')
        csv_writer.writerow(cols)
        
    
finally: 
    driver.quit()
    
csv_file.close()
```
